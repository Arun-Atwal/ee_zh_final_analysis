```python
import matplotlib.pyplot as plt
import numpy as np
import mplhep as hep
import csv

# hep.style.use(hep.style.ROOT)
hep.style.use("ATLAS")
```

```python
"""Delta phi, chbtil"""
chbtil = []
q = []
a = []
cl = []
lower_error = []
upper_error = []
with open('data/delta_phi_chbtil.csv') as csvfile:
    reader = csv.reader(csvfile)
    next(reader)
    for row in reader:
        chbtil.append(float(row[0]))
        q.append(float(row[1]))
        a.append(float(row[2]))
        cl.append(float(row[3]))
        lower_error.append(float(row[4]))
        upper_error.append(float(row[5]))
```

```python
# plt.plot(chbtil, a)
# plt.plot(chbtil, cl, '.--')
plt.plot(chbtil, [1-i for i in a], 'darkgreen', label = r'1-$p$')
plt.plot(chbtil, [1-i for i in cl], 's:', color = 'tab:gray', label = r'1-$CL_s$', linewidth = 1, markersize = 5)
plt.errorbar(chbtil, [1-i for i in cl], yerr = [upper_error, lower_error], fmt = 'none', color = 'tab:gray',
             elinewidth=1, capsize=3)
plt.yscale('linear')
plt.axhline(y=0.95, color='tab:red', linestyle='--', linewidth = 1)
plt.ylim(0,1)
plt.xlim(0, 35)
plt.legend(fontsize = 18, loc = 'lower right')
plt.xlabel(r'$c_{H\widetilde{B}} / \Lambda^2$')
plt.ylabel('Confidence')
plt.text(0.2, 0.91, r'1-$p$ = 0.95, 95% CL: [-31,31]', fontsize = 16, color = 'tab:red')
plt.title(r'Confidence across $c_{H\widetilde{B}} / \Lambda^2$ for $\Delta \phi_{\mu \mu}$')
plt.savefig('pcls.pdf')
```

```python
for i in chbtil[1:]:
    chbtil.insert(0, -i)
for i in q[1:]:
    q.insert(0, i)
for i in a[1:]:
    a.insert(0, i)
for i in cl[1:]:
    cl.insert(0, i)
```

```python
plt.plot(chbtil, q, 'sk-', markersize = 3)
plt.axhline(y=3.8415, color='tab:red', linestyle='--', linewidth=1)
plt.axvline(x=-31, color='k', linestyle='--', linewidth=1)
plt.axvline(x=31, color='k', linestyle='--', linewidth=1)
plt.yscale('linear')
plt.xlim(-35, 35)
# plt.ylim(0, 5)
plt.xlabel(r'$c_{H\widetilde{B}} / \Lambda^2$')
plt.ylabel(r'$q$')
plt.text(-15, 4, r'$q$ = 3.8415, 95% CL: [-31,31]', fontsize = 16, color = 'tab:red')
plt.title(r'Test statistic across $c_{H\widetilde{B}} / \Lambda^2$ for $\Delta \phi_{\mu \mu}$')
plt.text(0.21, 0.2, r'$\sqrt{s} = 0.24$ TeV, $1.7 \, \text{ab}^{-1}$', fontsize=16, transform=plt.gcf().transFigure)
plt.text(0.55, 0.2, r'$e^+ e^- \rightarrow (Z h) \rightarrow \mu^+ \mu^- \tau^+ \tau^-$', fontsize=16, transform=plt.gcf().transFigure)
plt.savefig('q.pdf')
```
