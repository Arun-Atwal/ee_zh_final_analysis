```python
import matplotlib.pyplot as plt
import numpy as np
import mplhep as hep
import csv

# hep.style.use(hep.style.ROOT)
hep.style.use("ATLAS")
```

```python
"""Delta phi, chbtil"""
chbtil = []
q = []
a = []
cl = []
lower_error = []
upper_error = []
with open('data/delta_phi_chbtil.csv') as csvfile:
    reader = csv.reader(csvfile)
    next(reader)
    for row in reader:
        chbtil.append(float(row[0]))
        q.append(float(row[1]))
        a.append(float(row[2]))
        cl.append(float(row[3]))
        lower_error.append(float(row[4]))
        upper_error.append(float(row[5]))
```

```python
# plt.plot(chbtil, a)
# plt.plot(chbtil, cl, '.--')
plt.plot(chbtil, [1-i for i in a], 'darkgreen', label = r'1-$p$')
plt.plot(chbtil, [1-i for i in cl], 's:', color = 'tab:gray', label = r'1-$CL_s$', linewidth = 1, markersize = 5)
plt.errorbar(chbtil, [1-i for i in cl], yerr = [upper_error, lower_error], fmt = 'none', color = 'tab:gray',
             elinewidth=1, capsize=3)
plt.yscale('linear')
plt.axhline(y=0.95, color='tab:red', linestyle='--', linewidth = 1)
plt.ylim(0,1)
plt.xlim(0, 35)
plt.legend(fontsize = 18, loc = 'lower right')
plt.show
```

```python
for i in chbtil[1:]:
    chbtil.insert(0, -i)
for i in q[1:]:
    q.insert(0, i)
for i in a[1:]:
    a.insert(0, i)
for i in cl[1:]:
    cl.insert(0, i)
```
