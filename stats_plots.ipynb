```python
import matplotlib.pyplot as plt
import numpy as np
import mplhep as hep
import csv

# hep.style.use(hep.style.ROOT)
hep.style.use("ATLAS")
```

```python
"""Delta phi, chbtil"""
chbtil = []
q = []
a = []
with open('data/delta_phi_chbtil.csv') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        chbtil.append(row[0])
        q.append(row[1])
        a.append(row[2])

plt.plot(chbtil[1:], q[1:])
plt.yscale('linear')
plt.show()
```

```python
plt.plot(chbtil[1:], a[1:])
plt.yscale('linear')
plt.show
```
